```python
nota1 = float(input("insira a primeira nota: "))
nota2 = float(input("insira a segunda nota: "))
nota3 = float(input("insira a terceira nota: "))

media = (nota1 + nota2 + nota3)/3

print("A media é %.3f"%(media))

if media >= 6:
    print("aprovado")
elif media <4:
    print("reprovado")
elif media>4 and media<6:
    print("recuperação")
else:
    print("outro")

```
