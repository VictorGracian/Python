```python
# Faça um programa que peça a quantidade de notas de um aluno, em seguida receba cada nota e apresente ao final a média
#do aluno informando informando se ele foi aprovado ou reprovado (media de aprovação =6)

qnt = int(input("quantos valores serao digitados"))
cont = 1
media = 0
soma = 0
situacao = ""
while cont <= qnt:
    soma += int(input("insira a nota"))
    cont +=1
    
media = soma/qnt
if(media>=6):
    situacao = "aprovado"
else:
    situacao = "reprovado"
print(media)
print(situacao)
```
